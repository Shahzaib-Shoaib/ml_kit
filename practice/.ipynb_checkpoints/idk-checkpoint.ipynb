```python
import pandas as pd
```

```python
df = pd.read_csv("train_data.csv")
```

```python
df.sample(15)
```

```python
#Number_of_Active_Investments:
df['Number_of_Active_Investments'] = df['Number_of_Active_Investments'].replace({'Single': 1, 'TRUE':1.0 ,  'Auto': 1.0 ,   'Marr':2.0 ,  '1':1.0,  '2':2.0 ,  '3':3.0, '4':4.0})
```

```python
#Investment_Duration  :
df['Investment_Duration'] = df['Investment_Duration'].replace({'36': 36 , '24':24 ,'60':60 , '12':12 ,'48': 48 ,'no': 0.0 , 'Ye':0.0 , "Master's":0.0})
```

```python
#Employment_Status  :
df['Employment_Status'] = df['Employment_Status'].replace({'Part':'Part-time' ,'Full' : 'Full-time' , 'Self':'Self-employed' , 'Unem':'Unemployed' , 'Unemloyed':'Unemployed' ,'1':'Full-time','Yes':'Full-time'  })
```

```python
#Marital_Status  :
df['Marital_Status'] = df['Marital_Status'].replace({'Div':'Divorced' , 'Sgl':'Single', 'Marr':'Married' , 'No':'Single'})
```

```python
#Owns_Property :
df['Owns_Property'] = df['Owns_Property'].replace({'Ye':'Yes' ,'yes':'Yes','no':'No' , 'NO':'No' , 'YES':'Yes' ,'Maybe':'Unknown'  , 'YE':'Yes', 'ye':'Yes' , 'Home':'Yes' })
```

```python
#Has_Dependents:
df['Has_Dependents'] = df['Has_Dependents'].replace({'Ye':'Yes' ,'yes':'Yes','no':'No' ,'ye':'Yes','YE':'Yes', 'NO':'No' , 'YES':'Yes' ,'Maybe':'Unknown' , 'TRUE':'Yes'})
```
